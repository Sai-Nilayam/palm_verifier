Objective
-------------
This project is verifying Palms of different persons. The idea behind this 
is, the way a Siamese network learns a function to find difference between 
two faces, can we train a Siamese network to find difference between two
Palm images.

Humen can differentiate difference faces. The same way they can also 
differentiate different Palms.

Iteration - 1
----------------
1. First of all two different pictures of palms of 8 different persons were taken and saved in the project folder at 'data/palm_images/'. The images are preprocessed from the 'raw_data' folder. The way the images are pre-processed is as follows:
    1. All images should have same orientation (portrait, finger-down, thumb on right)
    2. All images should have same resolution. 

2. All the ideces were written on a paper, pictured and stored in the projdect folder at 'data/'. 

3. All indices were saved to a csv file names as 'indices.csv' at 'data/'.

4. All the Palm images were renamed as 'x_y.jpg', x->person_index, y->image_number.
